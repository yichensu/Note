```python
import csv
```

```python
contents = []

with open('./台灣.csv', newline='') as csvfile:
  rows = csv.reader(csvfile)
  for row in rows:
    contents.append(row[0])
```

```python
print(len(contents))
contents
```

```python
def dic_gen(x):
    import re #（去掉各種詞）
#     content = all_sheet.range(f"A{x}").value
    content = contents[x]
    r1 = u"[^\u4e00-\u9fa5]"#去除英文及符號
    content_clean = re.sub(r1, '', content)
    content_clean #去完後狀況
    
    dic = {} #跑裡面的字
    gram2 = [] #跑完放進list
    for i in range(len(content_clean)-1):
        gram2.append(content_clean[i] + content_clean[i+1])
    gram2_unique = set(gram2) #用set去掉重複的詞
    gram2_unique #做成dict，key是詞，value是出現幾次
    for element in gram2_unique:
        dic.setdefault(element, gram2.count(element))
    return(dic)

dic_gen(2)
```

```python
dic2_all = {} #有多個dict，會逐一比較，若有類似的詞彙相加
for row in range(len(contents)):
    for k,v in dic_gen(row).items():
        if k in dic2_all.keys():
            dic2_all[k] += v
        else:
            dic2_all[k] = v
            
for key in dic2_all.copy(): #篩除詞頻<=50
    if dic2_all[key] <= 50:
        dic2_all.pop(key)
```

```python
dic2_all
```

```python
dic2_df = {} #有在內容出現過一次就加一
for key in dic2_all.keys(): #是否有機會是小tag，可以辨認在某篇文章中（文章數）
    count = 0
    for x in range(len(contents)):
#         content = all_sheet.range(f"A{x}").value
        content = contents[x]
        if key in content:
            count += 1
#           df_count = count
    dic2_df.setdefault(key, count)
```

```python
dic2_df
```

```python
sorted(dic2_df.items(), key=lambda d: d[1], reverse=True)
```

```python
stopword = []

with open ('./stopWords.txt', 'r', encoding = 'utf8') as txtfile: #讀stopword
    for data in txtfile.readlines():
        data = data.strip()
        stopword.append(data)


stopword
```

```python
for j in list(dic2_df.keys()):
    for l in range(len(stopword)):
        stop = stopword[l]
        if j in stop:
            print(j) #測試用，正式的可刪除
            dic2_df.pop(j)
            break
```
